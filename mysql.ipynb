```python
import pandas as pd
import mysql.connector
```

```python
# Đường dẫn đến file Excel của bạn
file_path = "./datacsv/news_vietnamnet.csv"  # Thay thế bằng đường dẫn thực tế

# Đọc file Excel (hoặc CSV) vào DataFrame
df = pd.read_csv(file_path)

# Fill na
df = df.fillna('')
print(len(df))
```

```python
df.head(10)
```

```python
# Tạo kết nối đến MySQL
connection = mysql.connector.connect(
    host="localhost",     # Thường là localhost
    user="root",          # Người dùng MySQL, mặc định là root
    password="",          # Mật khẩu MySQL, để trống nếu không có
    database="pbl6"  # Tên cơ sở dữ liệu
)
```

```python
# Tạo một con trỏ (cursor) để thực hiện truy vấn
cursor = connection.cursor()
```

```python
# Vòng lặp qua từng hàng trong DataFrame và lưu vào database
for index, row in df.iterrows():
    # Câu lệnh SQL để chèn dữ liệu vào bảng
    sql = """
    INSERT INTO tbl_news (Title, Content, Source)
    VALUES (%s, %s, %s)
    """
    
    # Dữ liệu tương ứng từ các cột trong DataFrame
    values = (row['Title'], row['Content'], row['Source'])
    
    # Thực thi câu lệnh SQL với dữ liệu tương ứng
    cursor.execute(sql, values)
```

```python
# Lưu thay đổi vào database
connection.commit()
```

```python
# Đóng con trỏ và kết nối
cursor.close()
connection.close()
```
